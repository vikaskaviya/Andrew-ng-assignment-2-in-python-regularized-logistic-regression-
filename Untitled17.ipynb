```python
 def sigmoid(z):
        return 1/(1+np.exp(-z))
```

```python
def cost(theta,X,y,learningrate):
    theta = np.matrix(theta)
    X = np.matrix(X)
    y = np.matrix(y)
    
    first = np.multiply(y,np.log(sigmoid(X*theta.T)))
    second = np.multiply((1-y),np.log(1-sigmoid(X*theta.T)))
    reg = (learningrate/2*len(y))*np.sum(np.power(theta[:,1:theta.shape[1]],2))
    
    return (-1*(np.sum(first+second)))/len(y)+ reg
    
    
    
```

```python
def gradient(theta,X,y,learningrate) :
    theta = np.matrix(theta)
    X = np.matrix(X)
    y = np.matrix(y)
    
    parameters = int(theta.ravel().shape[1])
    print(parameters)
    grad = np.zeros(parameters)
    
    error = sigmoid(X*theta.T) - y
   
    for i in range(parameters) :
        term = np.multiply(error, X[:,i])
        
        if(i==0):
            grad[i] = np.sum(term)/len(y)
        else :
            grad[i] = np.sum(term)/len(y) + ((learningrate/len(y))*theta[:,i])
    
    return grad
        
```

```python
col = data.shape[1]
print(col)
data.head()
```

```python
X=data.iloc[:,0:col-1]
y = data.iloc[:,col-1:col]



print(X.shape[1])
print(y.shape[1])
theta = np.zeros(28)
print(X.shape,theta.shape,y.shape)

X = np.array(X.values)
y = np.array(y.values)

print(X.shape,theta.shape,y.shape)
print(theta)
learningrate = 1
print(cost(theta,X,y,learningrate))
```

```python
def predict(theta_min,X):
    prob = sigmoid(X*theta_min.T)
    return [1 if X >= 0.5 else 0 for X in prob]
```

```python
theta_min = np.matrix(theta)
print(theta_min)

predictions = predict(theta_min, X)

count = 0
for (a,b) in zip(predictions,y) :
    if (a == 1 and b==1) or (a==0 and b == 0):
        count = count+1


accuracy = (count/118)*100
print(accuracy)
```

```python
import scipy.optimize as opt
result = opt.fmin(cost, x0=theta, args = (X,y,learningrate),full_output = True)
print(result[0])
print(result[1])

print(cost(result[0],X,y,learningrate))
```

```python
import scipy.optimize as opt
result = opt.fmin_tnc(cost, x0=theta,fprime = gradient, args = (X,y,learningrate))
print(result[0])
print(result[1])

print(cost(result[0],X,y,learningrate))
```

```python
theta_min = np.matrix(result[0])
print(theta_min)

print(cost(theta_min,X,y,learningrate))
```
